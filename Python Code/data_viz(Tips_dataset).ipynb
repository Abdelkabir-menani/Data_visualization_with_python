```python
s=pd.DataFrame(data['sex'].value_counts())
s.reset_index(inplace=True)
r='Male','Female'
s.plot(kind='pie',y='sex',labels=r,autopct='%1.2f')
plt.show()
```

```python
t=pd.DataFrame(data['Fumeur'].value_counts())
t.reset_index(inplace=True)
plt.bar(t['index'],t['Fumeur'])
print(t)
```

```python
print(data['Jour_semaine'].unique())
```

```python
g = sns.catplot(x="Jour_semaine",kind='count',order=['Jeudi','Vendredi','Samedi','Dimanche'],data=data);
print(g.fig.suptitle("Fréquence des jours de la semaine dans notre dataset[Count Plot]",y=1.05))
```

```python
m=pd.DataFrame(data['soir_jour'].value_counts())
m.reset_index(inplace=True)
plt.bar(m['index'],m['soir_jour'])
print(m)
```

```python
print(data['nb_personne'].unique())
```

```python
f = sns.catplot(x="nb_personne",kind='count',order=[1,2,3,4,5,6],data=data);
print(f.fig.suptitle("Fréquence des nombres des personnes qui consomment dans notre dataset[Count Plot]",y=1.05))
```

```python
g = sns.catplot(x="Jour_semaine",y="Pourboire",kind='bar',order=['Jeudi','Vendredi','Samedi','Dimanche'],ci=False,data=data);
print(g.fig.suptitle('Le nombre des pourboires par le jour de la semaine[Bar Plot]',y=1.05))
```

```python
g = sns.catplot(x="Jour_semaine",y="Pourboire",kind='strip',order=["Jeudi",'Vendredi','Samedi','Dimanche'],ci=False,data=data);
print(g.fig.suptitle('La somme des pourboires par le jour de la semaine[Strip Plot]',y=1.03))
```
